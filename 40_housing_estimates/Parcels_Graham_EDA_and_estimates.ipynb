# Generating estimates
I start by making sure all units in the same tract have the same total housing population. Then I create a new column for "estimated_housing_pop2" that contains the number of etimate housing units per row. This will be 1 for almost all of the columns except for duplex/triplex, where it will be 2. A second column will be created, "estimated_housing_pop2" where duplex/triplex will be counted as 3. I will then group by for both columns to get their total housing units per tract, for each counting method. Finally two more columns will be created which will find the difference between the estimates and the estimates I calculated.

```python
# checking the number of "Duplex/triplex" rows
duplex_check = graham_final[graham_final["PARUSEDESC"] == "Duplex/triplex"]
duplex_check
```

```python
# creating estimate columns with all values as 1
graham_final["estimated_housing_pop2"] = 1
graham_final["estimated_housing_pop3"] = 1
```

```python
# generating the column "estimated_housing_pop2"
# if "PARUSEDESC" is "Duplex/triplex" then "estimated_housing_pop2" is 2
import numpy as np

graham_final["estimated_housing_pop2"]= np.where(graham_final["PARUSEDESC"] == "Duplex/triplex", 2, 1)
graham_final["estimated_housing_pop3"]= np.where(graham_final["PARUSEDESC"] == "Duplex/triplex", 3, 1)
graham_final
```

### Groupby Tract
Now that the columns for the estimates are created I will groupy the tract to generate the total housing population estimates for each tract. The first groupby will be for the method where duplex/triplex is counted as 2 and the second is for the method where duplex/triplex is counted as 3.

```python
# groupby full_fips for "estimated_housing_pop2" and "estimated_housing_pop3" sums per tract
graham_final['total_estimated_housing_pop2'] = graham_final.groupby("full_fips", sort=False)["estimated_housing_pop2"].transform('sum')
graham_final['total_estimated_housing_pop3'] = graham_final.groupby("full_fips", sort=False)["estimated_housing_pop3"].transform('sum')
graham_final.head()
```

### Percent Difference
Calculating the percent differnce between our estimates and the total housing population counts. There will be two different percent differences, "housing_per_diff2" and "housing_per_diff3". The first diff is for the method with counting duplex/triplex as 2, while the second diff is for counting them as 3.

```python
# converting all populations and estimates to floats
graham_final["total_housing"]= graham_final["total_housing"].astype(float)
graham_final["total_estimated_housing_pop2"]= graham_final["total_estimated_housing_pop2"].astype(float)
graham_final["total_estimated_housing_pop3"]= graham_final["total_estimated_housing_pop3"].astype(float)
```

```python
# creating columns for percent difference
graham_final["housing_per_diff2"] = ((graham_final["total_estimated_housing_pop2"] - graham_final["total_housing"])/graham_final["total_housing"])*100

graham_final["housing_per_diff3"] = ((graham_final["total_estimated_housing_pop3"] - graham_final["total_housing"])/graham_final["total_housing"])*100

graham_final.head()
```

```python
# seeing percent diff per tract by removing duplicates
per_diff_check = graham_final.drop_duplicates(subset=['full_fips'], keep='first')
per_diff_check
```

```python
#grouping by tract

import pandas as pd
perc_diff_df= pd.DataFrame(graham_final[["full_fips","total_housing", "total_estimated_housing_pop3", "housing_per_diff2", "housing_per_diff3"]])

# seeing percent diff per tract by removing duplicates
per_diff_check = perc_diff_df.drop_duplicates(subset=['full_fips'], keep='first')
per_diff_check
```
